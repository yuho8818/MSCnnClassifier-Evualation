```python
import os
import re
filename = []
for root, dirs, files in os.walk('./carNeuralNet/images'): 
    filename.append(files)
print(filename)
```

```python
len(filename[0])
```

```python
import pandas as pd
index = [i for i in range(len(filename[0]))]
df = pd.DataFrame({'Image Index':filename[0]})
```

```python
df.head()
```

```python
df['brand'] = df['Image Index'].apply(lambda x:''.join(re.findall('\D',x)[:-11]))
```

```python
df.head()
```

```python
df.tail()
```

```python
brand = set()
df['brand'].apply(lambda x: brand.add(x))
```

```python
brand = list(brand)
brand.sort()
brand
```

```python
for i in brand:
    df[i] = 0
    df[i][df.brand==i] = 1 
```

```python
df[df['Alfa Rome']==1]
```

```python
train_data = df.sample(frac=0.7,random_state=0,axis=0)
print(train_data.shape)
remove_train_df = df[~df.index.isin(train_data.index)]
print(remove_train_df.shape)
test_df = remove_train_df.sample(frac=0.666667,random_state=0,axis=0)
print(test_df.shape)
dev_df = remove_train_df[~remove_train_df.index.isin(test_df.index)]
dev_df = dev_df.sample(frac=1,random_state=0,axis=0)
print(dev_df.shape)
```
